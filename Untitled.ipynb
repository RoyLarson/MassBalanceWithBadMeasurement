```python
%load_ext autoreload
import padasip as pa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import filterpy as fp
from collections import namedtuple
%matplotlib inline
%autoreload
```

```python
data_file = pd.ExcelFile("hdata.xlsx")
```

```python
columns = ['Date', 'Time', 'AM_PM', 'Outlet', 'Inlet_1', 'Inlet_2', 'Inlet_3', 'Inlet_4']
raw_data = pd.read_excel(data_file, header =None, names = columns, usecols = 'B, C, D, F, I, L, O, R')
```

```python
raw_data.head()
```

```python
datetime = pd.to_datetime(raw_data.Date.astype(str).str.strip()+' '+raw_data.Time.astype(str).str.strip()+' '+raw_data.AM_PM.str.strip())
datetime.name = 'DateTime'
```

```python
data = raw_data.loc[:,'Outlet':'Inlet_4'].join(datetime)
```

```python
data.set_index('DateTime', inplace = True)
clean_data = data.dropna(how = 'any')
clean_data = clean_data[clean_data.applymap(np.isreal).all(1)]
clean_data = clean_data.infer_objects()
interp_data = clean_data.resample('2 min').mean()
interp_data = interp_data.interpolate()
interp_data = interp_data.fillna('ffill')
interp_data = interp_data.fillna('bfill')
interp_data.to_csv('data.csv')
```

```python
f = pa.filters.AdaptiveFilter(model = 'NLMS', n = 4, mu = .01, w='random')
```

```python
Result = namedtuple('Result',['Est', 'Err', 'W', 'ELBND'] )
def all_comb(f, data):

    results = dict()
    columns = data.columns
    for col in columns:
        cols = columns.tolist()
        cols.remove(col)
        est, err, w = f.run(data.loc[:,col].values, data.loc[:,cols].values)
        w = pd.DataFrame(w, columns = cols, index = data.index)
        elbnd = pa.detection.ELBND(w, err, function = 'sum')
        results[col] = Result(est, err, w, elbnd)
    return results
```

```python
results = all_comb(f, interp_data)

#Bulk_Oil_est, Bulk_Oil_Err, Bulk_Oil_w = f.run(interp_data.HBlk_Oil.values, interp_data.loc[:,'H18H_Oil':].values)
#H18H_Oil_est, H_18H_Oil_Err, H_18H_w = f.run(iterp_data.H18H_Oil.values, interp_data.loc[:,['HBlk_Oil', ''])
```

```python
num_results = len(results)
fig_est = plt.figure(figsize=(15,10*num_results))
plt.xlabel("samples - k")
plt.title("Adapation and Error")
plt.ylim(0,4000)
i = 0
for meas, result in results.items():
    i+=1
    ax = plt.subplot(num_results, 1, i)
    ax.set_ylim(0,4000)
    ax.plot(interp_data[meas].values,"b", label=f"{meas} act: target")
    ax.plot(result.Est,"g", label=f"{meas} est: output", alpha = .5);
    ax.legend()


plt.tight_layout();
plt.show();
```

```python
fig_est = plt.figure(figsize=(15,10*num_results))
plt.xlabel("samples - k")
plt.title("Error")

i = 0
for meas, result in results.items():
    i+=1
    ax = plt.subplot(num_results, 1, i)
    ax.plot(10*np.log10(result.Err**2),color = 'b', label=f"{meas} err: dB");
    ax.plot(10*np.log10(result.ELBND**2), color = 'g', label=f'{meas} elbnd: dB', alpha = .7);
    ax.legend()


plt.tight_layout();
plt.show();
```

```python
fig_est = plt.figure(figsize=(15,10*num_results))
plt.xlabel("samples - k")
plt.title("Weights")
plt.ylim(0,4000)
i = 0
for meas, result in results.items():
    i+=1
    ax = plt.subplot(num_results, 1, i)
    ax.set_ylim(0,4000)
    ax.plot(interp_data[meas].values,"b", label=f"{meas} act: target")
    ax.plot(result.Est,"g", label=f"{meas} est: output", alpha = .5);
    ax.legend()


plt.tight_layout();
plt.show();
```

```python
affine = pa.filters.FilterAP(n=4, order=5, mu=0.05, eps=0.001, w=[1,1,1,1])
y, e, w = affine.run(interp_data.HBlk_Oil.values, interp_data.loc[:,'H18H_Oil':].values)
elbnd = pa.detection.ELBND(w, e, function = 'max')
learn_entropy = pa.detection.learning_entropy(w, m=30, order=1)
weights = pd.DataFrame(w, columns = interp_data.columns[1:], index = interp_data.index)
weights.rolling(window = 600).mean().plot(figsize = (15,10), logy = True)
```
